```python
# 백준 1874

n = int(input())

count = 1
stack = []
result = []

for i in range(1, n+1): # 데이터 개수만큼 반복
    data = int(input())
    while count <= data: # 입력 받은 데이터에 도달할 때까지 삽입
        stack.append(count)
        count += 1
        result.append('+')
    if stack[-1] == data: # 최상위 데이터 원소가 데이터과 같을 때 빼기 실행
        stack.pop()
        result.append('-')
    else: # 불가능한 경우
        print('NO')
        exit(0)

for i in result:
    print(i)
```

```python
# 백준 17928 시간초과 X

n = int(input()) # 수의 개수
arr = list(map(int, input().split())) # 수열 데이터
stack = [] # 스택(stack) 초기화
NGE = [-1] * n # 오큰수 배열'

for i in range(n):
    
    x = arr[i] # 하나씩 수 확인
    
    if len(stack) == 0 or stack[-1][0] >= x: # 내림차순 형태라면(작거나 같은 원소를 만났다면)
        stack.append((x, i)) # (수, 인덱스) 형태로 삽입
        
    else: # 오름차순 형태라면(큰 수를 만났다면)
        while len(stack) > 0: # 역방향으로 하나씩 꺼내기
            previous, index = stack.pop()

            if previous >= x: # 크거나 같은 이전 원소를 만났다면 다시 삽입
                stack.append((previous, index))
                break
            else:
                NGE[index] = x # 오큰수 기록

        stack.append((x, i)) # (수, 인덱스) 형태로 삽입

for x in NGE: # 오큰수를 하나씩 출력
    print(x, end=' ')
```

```python
# 백준 17413 (구현 + 스택)

S, tmp = input(), ""

ck = False
ans = ''

for i in S:
    if i == ' ':
        if not ck: # <> 밖
            ans += tmp[::-1] + ' ' # 지금까지 저장한걸 반대로 덧 붙히고 ' ' 랑 tmp 비워주기
            tmp = ''
        else:
            ans += ' '
    elif i in '<':
        ck = True # 괄호안 시작
        ans += tmp[::-1] + '<'
        tmp = '' # tmp 비워주기
    elif i in '>':
        ck = False # 괄호 끝
        ans += '>' # 괄호로 닫아주기
    else: # 알파벳과 숫자
        if ck: # 괄호 안이라면 그대로, 괄호 밖이라면 tmp에 덧 붙혀줌
            ans += i
        else:
            tmp += i
            
ans += tmp[::-1]

print(ans)
```

```python
# 백준 1874 스택

# 다른 풀이 : 숫자를 입력 받을때마다 top과 비교 data공간이 0이거나 받은숫자가 top보다 크다면 삽입
#                    같다면 pop , top보다 n이 더 작다면 no

# 8
# 4
# 3
# 6
# 8
# 7
# 5
# 2
# 1

# + + + + - - + + - + + - - - - - 
s = []
n = int(input())
number = [i for i in range(1, n+1)] # 1~n
stack = []
for _ in range(n):
    s.append(int(input()))

stack.append(number.pop(0))

rst = []
rst.append('+')
check = True

while number: 
    # s [4, 3, 6, 8, 7, 5, 2, 1]  
    # n [2, 3, 4, 5, 6, 7, 8]  
    # stack [1] 4 1 비교 -> 다름 pu stack[1, 2] -> 다름 pu stack[1, 2, 3] -> 다름 pu stack[1, 2, 3, 4] 비교 같음 -> s[0]을 pop

    if not stack or stack[-1] != s[0]:
        stack.append(number.pop(0))
        rst.append('+')
    elif stack[-1] == s[0]:
        s.pop(0)
        stack.pop()
        rst.append('-')
    
while stack:
    if s[0] == stack[-1]:
        s.pop(0)
        stack.pop()
        rst.append('-')
    else:
        check = False
        break
        
if check:
    for i in rst:
        print(i)
else:
    print('NO')
    
    
```

```python
# 백준 1406 스택 -> PyPy로 제출해야 시간초과 안걸림 (insert, pop를 앞쪽에 가할 경우 새로운 리스트를 생성해야 하므로 시간초과)

# abcd
# 3
# P x
# L
# P y

# abcdyx

# L   	커서를 왼쪽으로 한 칸 옮김 (커서가 문장의 맨 앞이면 무시됨)
# D     커서를 오른쪽으로 한 칸 옮김 (커서가 문장의 맨 뒤이면 무시됨)
# B     커서 왼쪽에 있는 문자를 삭제함 (커서가 문장의 맨 앞이면 무시됨)
#       삭제로 인해 커서는 한 칸 왼쪽으로 이동한 것처럼 나타나지만, 실제로 커서의 오른쪽에 있던 문자는 그대로임
# P $   $라는 문자를 커서 왼쪽에 추가함

stack_l = [i for i in input()]
stack_r = []

for _ in range(int(input())):
    C = list(input().split()) # c[0] 명령어 c[1]은 P만있음

    if C[0] == 'L':
        if not stack_l:# 커서가 맨앞이면 무시
            continue
        stack_r.append(stack_l.pop()) # 앞쪽에 insert시 시간초과 
    elif C[0] == 'D':
        if not stack_r:# 커서가 맨뒤면 무시
            continue
        stack_l.append(stack_r.pop())
    elif C[0] == 'B':
        if not stack_l:# 커서가 맨앞이면 무시
            continue
        stack_l.pop()
    else:
        stack_l.append(C[1])

rst = stack_l + stack_r[::-1]
print(''.join(rst))
```

```python
# 백준 17299 스택

n = int(input())
A = list(map(int, input().split()))

rst = {}
for i in A:
    if i not in rst:
        rst[i] = 1
    else:
        rst[i] += 1

B = [rst[i] for i in A] # 3 3 2 1 1 2 3  등장횟수 삽입
answer = [-1] * n
stack = []
stack.append(0) # index를 통해 비교를 하기에 처음에 0을 삽입

# 아래 for문에서 못구해도 처음 초기화에서 -1을 넣어주었기에 그대로 -1로 출력이 됨
for i in range(1, n):
    while stack and B[stack[-1]] < B[i]:
        answer[stack.pop()] = A[i]
    stack.append(i)

print(*answer)
```
