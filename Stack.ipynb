```python
# 백준 1874

n = int(input())

count = 1
stack = []
result = []

for i in range(1, n+1): # 데이터 개수만큼 반복
    data = int(input())
    while count <= data: # 입력 받은 데이터에 도달할 때까지 삽입
        stack.append(count)
        count += 1
        result.append('+')
    if stack[-1] == data: # 최상위 데이터 원소가 데이터과 같을 때 빼기 실행
        stack.pop()
        result.append('-')
    else: # 불가능한 경우
        print('NO')
        exit(0)

for i in result:
    print(i)
```

```python
# 백준 5397
test_case = int(input())

for _ in range(test_case):
    str_ = list(input())
    stack_l = list()
    stack_r = list()
    for i in str_:
        if i == '<':
            if stack_l:
                stack_r.append(stack_l.pop())
        elif i == '>':
            if stack_r:
                stack_l.append(stack_r.pop())
        elif i == '-':
            if stack_l: # 있는지판단을 해야함 없는데 pop을 하면 에러 발생
                stack_l.pop()
        else:
            stack_l.append(i)
    stack_l.extend(reversed(stack_r)) # stack_r은 역순으로 해야 원래 패스워드로 붙혀줘야함
    print(''.join(stack_l))
```

```python
# 백준 10828
import sys
input = sys.stdin.readline  # 빠른 입력

test_case = int(input())
stack = []

for _ in range(test_case):
    N_list = input().strip().split(' ') # list(map(str, input().split())) 사용시 시간초과
    
    if N_list[0] == 'push':
        stack.append(N_list[1])
    elif N_list[0] == 'pop':
        if stack:
            print(stack.pop())
        else:
            print(-1)
    elif N_list[0] == 'size':
        print(len(stack))
    elif N_list[0] == 'empty':
        if len(stack) == 0:
            print(1)
        else:
            print(0)
    elif N_list[0] == 'top':
        if len(stack) == 0:
            print(-1)
        else:
            print(stack[-1])
```

```python
# 백준 1874

test_case = int(input())
result = [] # 최종 정답
data = []   # 스택 초기화
count = 0 # 삽입할 수

for _ in range(test_case):
    n = int(input()) # 하나씩 입력
    # top()보다 n이 더 크다면, 스택에 삽입
    while len(data) ==0 or n > data[-1]:
        count += 1
        data.append(count)
        result.append('+')
    
    # top() 과 n가 같다면, 스택에서 제거
    if data[-1] == n:
        result.append('-')
        data.pop()
    # top()보다 n가 더 작다면, 불가능
    else:
        result = ['NO']
        break

for x in result:
    print(x)
```

```python
# 백준 17928 시간초과 O

test_case = int(input())
stack = list(map(int, input().split()))

result = []  # 정답
for j in range(1, test_case):
    data_1 = []
    for i in range(j, test_case):
        
        n = stack[j - 1]
        if n < stack[i]:
            
            data_1.append(stack[i])
        else:
            pass
    
    if len(data_1) == 0:
        result.append(-1)
    else:
        result.append(data_1[0])
    
result.append(-1)
for i in result:
    print(i, end=' ')
```

```python
# 백준 17928 시간초과 X

n = int(input()) # 수의 개수
arr = list(map(int, input().split())) # 수열 데이터
stack = [] # 스택(stack) 초기화
NGE = [-1] * n # 오큰수 배열'

for i in range(n):
    
    x = arr[i] # 하나씩 수 확인
    
    if len(stack) == 0 or stack[-1][0] >= x: # 내림차순 형태라면(작거나 같은 원소를 만났다면)
        stack.append((x, i)) # (수, 인덱스) 형태로 삽입
        
    else: # 오름차순 형태라면(큰 수를 만났다면)
        while len(stack) > 0: # 역방향으로 하나씩 꺼내기
            previous, index = stack.pop()

            if previous >= x: # 크거나 같은 이전 원소를 만났다면 다시 삽입
                stack.append((previous, index))
                break
            else:
                NGE[index] = x # 오큰수 기록

        stack.append((x, i)) # (수, 인덱스) 형태로 삽입

for x in NGE: # 오큰수를 하나씩 출력
    print(x, end=' ')
```

```python
# 백준 17413 (구현 + 스택)

S, tmp = input(), ""

ck = False
ans = ''

for i in S:
    if i == ' ':
        if not ck: # <> 밖
            ans += tmp[::-1] + ' ' # 지금까지 저장한걸 반대로 덧 붙히고 ' ' 랑 tmp 비워주기
            tmp = ''
        else:
            ans += ' '
    elif i in '<':
        ck = True # 괄호안 시작
        ans += tmp[::-1] + '<'
        tmp = '' # tmp 비워주기
    elif i in '>':
        ck = False # 괄호 끝
        ans += '>' # 괄호로 닫아주기
    else: # 알파벳과 숫자
        if ck: # 괄호 안이라면 그대로, 괄호 밖이라면 tmp에 덧 붙혀줌
            ans += i
        else:
            tmp += i
            
ans += tmp[::-1]

print(ans)
```
