```python
# 백준 14395(bfs)
from collections import deque

s, t = map(int, input().split()) # 시작 끝

queue = deque([])
queue.append((s, '')) # 시작점, 거리 0 추가
visited = set()
found = False
rst = []

while len(queue) != 0: # 큐가 빌 때까지 반복
    if s == t: # 처음부터 같은 경우
        found = True 
        print(0)
        break
    value, oper_ = queue.popleft() # 값과 거리 빼기
    if value > int(1e9): # 값의 범위를 벗어나는 경우
        continue
    if value == 0: # 0인 경우 제외
        continue
    if value == t: # 값을 찾은 경우
        found = True
        rst.append(oper_)
        break
    for oper in ['*', '+', '-', '/']:
        next_value = value
        if oper == '*':
            next_value *= next_value
        if oper == '+':
            next_value += next_value
        if oper == '-':
            next_value -= next_value
        if oper == '/':
            next_value /= next_value
        if next_value not in visited:
            queue.append((next_value, oper_ + oper)) # 값과 함께 거리를 queue에추가
            visited.add(next_value) # 방문 후 추가
            
if not found:
    print(-1)
    
rst.sort()
if rst:
    print(rst[0])
```

```python
# 백준 16768

def new_array(n): # 방문 체크 함수
    return [[False for i in range(10)] for _ in range(n)]

n, k = map(int, input().split())
M = [list(input()) for _ in range(n)]
ck = new_array(n)
ck2 = new_array(n)

dx, dy = [1, -1, 0, 0], [0, 0, -1, 1]

def dfs(x, y): # 탐색하는 함수
    ck[x][y] = True
    ret = 1
    for i in range(4):
        xx, yy = x + dx[i], y + dy[i]
        if xx <0 or xx >= n or yy < 0 or yy >= 10:
            continue
        if ck[xx][yy] or M[x][y] != M[xx][yy]: # 다르면 연결 x ex) 현재 값 : 1 / 다른값 : 2
            continue
        ret += dfs(xx, yy)
    return ret
        
def dfs2(x, y, val): # 지우는 함수
    ck2[x][y] = True
    M[x][y] = '0'
    for i in range(4):
        xx, yy = x + dx[i], y + dy[i]
        if xx < 0 or xx >= n or yy < 0 or yy >= 10:
            continue
        if ck2[xx][yy] or M[xx][yy] != val: # M[xx][yy] 값이 M[i][j] 값과 다르면 지우지않음
            continue
        dfs2(xx, yy, val)
        
def down(): # 내리는 함수
    for i in range(10):
        tmp = []
        for j in range(n):
            if M[j][i] != '0':
                tmp.append(M[j][i])
        for j in range(n-len(tmp)):
            M[j][i] = '0'
        for j in range(n-len(tmp), n):
            M[j][i] = tmp[j-(n-len(tmp))]

while True:
    exist = False
    ck = new_array(n)
    ck2 = new_array(n)
    for i in range(n):
        for j in range(10):
            if M[i][j] == '0' or ck[i][j] :
                continue
            ret = dfs(i, j) # 개수세기
            if ret >= k:
                dfs2(i, j, M[i][j]) # 지우기
                exist = True
                
    if not exist:
        break
    down() # 제거 하고 내려오는 함수
    
for i in M:
    print(''.join(i))
```
