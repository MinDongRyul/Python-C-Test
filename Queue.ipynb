```python
# 백준 1021(deque)
from collections import deque
n, m = map(int, input().split()) # 원소의 개수, 뽑아내는 횟수
d = deque([i for i in range(1, n + 1)]) # 1부터 N까지의 원소를 삽입
targets = list(map(int, input().split())) # 뽑아낼 원소 목록

cnt = 0 # 회전 연산 수행 횟수
for target in targets: # 뽑아낼 원소를 하나씩 확인하며
    
    index = d.index(target) # 덱에서 해당 원소의 위치를 찾기
    
    if index <= len(d) // 2: # 왼쪽으로 돌리는 게 더 빠른 경우
        for i in range(index): # 회전 연산 반복 수행
            x = d.popleft()
            d.append(x)
            cnt += 1
    else: # 오른쪽으로 돌리는 게 더 빠른 경우
        for i in range(len(d) - index): # 회전 연산 반복 수행
            x = d.pop()         # 맨 오른쪽 꺼내기
            d.appendleft(x) # 맨 왼쪽에 삽입
            cnt += 1
            
    d.popleft() # 원소 꺼내기
    
print(cnt) # 결과 출력
```

```python
# 백준 2346
from collections import deque

n = int(input()) # 원소의 개수 N
# 전체 원소 리스트
arr = list(map(int, input().split()))
d = deque() # 덱(deque) 초기화

for i in range(n):
    # (수, 번호) 형태로 원소를 삽입
    d.append((arr[i], i + 1)) # (3, 1), (2, 2), (1, 3), (-3, 4), (-1, 5)
    
result = [] # 결과 배열
current, index = d.popleft() # 처음 원소 추출 (3, 1),

result.append(index) # 1

for i in range(n - 1): # 원소를 모두 꺼내기
    
    if current > 0: # 양수라면
        # current - 1번 "왼쪽으로 돌리기" 수행
        for j in range(current - 1):
            x = d.popleft()
            d.append(x)
            
    else: # 음수라면 (0은 없음)
        # |current|번 "오른쪽으로 돌리기" 수행
        for j in range(-current):
            x = d.pop()
            d.appendleft(x)
            
    # 원소 추출
    current, index = d.popleft() # (-3, 4), (-1 , 5)
    result.append(index) # 4, 5

for x in result: # 결과 출력
    print(x, end=' ')
```

```python
# 백준 7662 (우선순위 큐)

import heapq

def pop(heap):
    while len(heap) > 0:
        data, id = heapq.heappop(heap)
        if not deleted[id]:
            deleted[id] = True
            return data
    return None

test_case = int(input())
for _ in range(test_case):
    heap = []          # 최소값 힙
    max_heap = [] # 최대값 힙
    current = 0
    
    test_case_2 = int(input())
    deleted = [False] * (test_case_2 + 1)
    
    for _ in range(test_case_2):
        k, n = list(map(str, input().split()))
        n = int(n)
        
        if k == 'I':
            heapq.heappush(heap, (n, current)) # 최소값 힙
            heapq.heappush(max_heap, (-n, current)) # 최대값 힙
            current += 1
        elif k == 'D':
            # 최대값 제거
            if n == 1:
                pop(max_heap) # 최대값 제거
            elif n == -1:
                pop(heap) # 최솟값 제거
    
    max_value = pop(max_heap)
    if max_value == None:
        print('EMPTY')
    else:
        heapq.heappush(heap, (-max_value, current))
        print(-max_value, pop(heap))
```

```python
# 백준 1766(위상 정렬 알고리즘 사용)

import heapq

n, m = map(int, input().split())
array = [[] for i in range(n + 1)]
indegree = [0] * (n + 1)
heap = []
result = []

for _ in range(m): # 입력을 받아 x에 연결된 y를 append
    x, y = map(int, input().split())
    array[x].append(y) # ex 4 2 array[4] = 2 -> 2에대해  간선 추가, 4 -> 2
    indegree[y] += 1 # 간선이 있으면 +=1
    
for i in range(1, n + 1):
    if indegree[i] == 0: # 간선이 없으면 heappush실행, 
        heapq.heappush(heap, i)
print(array)
result = []
while heap:
    data = heapq.heappop(heap) # 맨 처음 값을 pop
    result.append(data) # 결과 리스트에 삽입
    for y in array[data]: # ex array[3] 에서 1이 나오고
        indegree[y] -= 1 # 간선을 제거
        if indegree[y] == 0: # 간선이 제거되면 heap에 추가
            heapq.heappush(heap, y)
            
for i in result:
    print(i, end=' ')
```

```python
# 백준 1158 (덱)
from collections import deque

# 7 3
# <3, 6, 2, 7, 5, 1, 4>

def yose(n, k, lst):
    rst = []
    while lst:
        for i in range(k-1): # 세번 째
            lst.append(lst.popleft())
        rst.append(lst.popleft())
    
    print('<',end='')
    for idx, c in enumerate(rst):
        if idx == n-1:
            print(str(c), end='')
        else:
            print(str(c) +', ', end='')
    print('>')
    
n, k = map(int, input().split())
lst = deque([i for i in range(1, n+1)])
yose(n, k, lst)
```

```python
# 백준 18258

# push X: 정수 X를 큐에 넣는 연산이다.
# pop: 큐에서 가장 앞에 있는 정수를 빼고, 그 수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.
# size: 큐에 들어있는 정수의 개수를 출력한다.
# empty: 큐가 비어있으면 1, 아니면 0을 출력한다.
# front: 큐의 가장 앞에 있는 정수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.
# back: 큐의 가장 뒤에 있는 정수를 출력한다. 만약 큐에 들어있는 정수가 없는 경우에는 -1을 출력한다.
from collections import deque

queue = deque([])
for _ in range(int(input())):
    n = input().split()
    
    if n[0] == 'push': # push 1
        queue.append(n[1]) # 1
    elif n[0] == 'pop': # pop
        if queue: # 큐 있음
            print(queue.popleft()) # deque로 하지 않으면 시간초과
        else:
            print(-1)
    elif n[0] == 'size': # size
        print(len(queue))
    elif n[0] == 'empty': # empty
        if queue: # 큐 있음
            print(0)
        else: # 큐 비어있음
            print(1)
    elif n[0] == 'front': # front
        if queue : # 큐 있음
            print(queue[0])
        else: # 큐 비어있음
            print(-1)
    elif n[0] == 'back':
        if queue : # 큐 있음
            print(queue[-1])
        else: # 큐 비어있음
            print(-1)
```

```python
# 백준 10866 덱

# push_front X: 정수 X를 덱의 앞에 넣는다.
# push_back X: 정수 X를 덱의 뒤에 넣는다.
# pop_front: 덱의 가장 앞에 있는 수를 빼고, 그 수를 출력한다. 만약, 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.
# pop_back: 덱의 가장 뒤에 있는 수를 빼고, 그 수를 출력한다. 만약, 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.
# size: 덱에 들어있는 정수의 개수를 출력한다.
# empty: 덱이 비어있으면 1을, 아니면 0을 출력한다.
# front: 덱의 가장 앞에 있는 정수를 출력한다. 만약 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.
# back: 덱의 가장 뒤에 있는 정수를 출력한다. 만약 덱에 들어있는 정수가 없는 경우에는 -1을 출력한다.

from collections import deque

queue = deque([])

for _ in range(int(input())):
    C = list(input().split())
    
    if C[0] == 'push_front':
        queue.insert(0, C[1]) # appendleft 로 하면 시간초과발생
    elif C[0] == 'push_back':
        queue.append(C[1])
    elif C[0] == 'pop_front':
        if not queue:
            print(-1)
            continue
        print(queue.popleft())
    elif C[0] == 'pop_back':
        if not queue:
            print(-1)
            continue
        print(queue.pop())
    elif C[0] == 'size':
        print(len(queue))
    elif C[0] == 'empty':
        if not queue: # 비어
            print(1)
        else:
            print(0)
    elif C[0] == 'front':
        if not queue:
            print(-1)
            continue
        print(queue[0])
    else:
        if not queue:
            print(-1)
            continue
        print(queue[-1])
```
